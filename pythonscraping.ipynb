```python
url = "https://stackoverflow.com/questions/tagged/python?sort=MostVotes&edited=true"
```

```python
from bs4 import BeautifulSoup
from splinter import Browser
```

```python
browser = Browser('chrome')
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html,'html.parser')
```

```python
question_div = soup.find_all('div',class_="s-post-summary--content")
print(question_div)
```

```python
question_list = []

for i in range(len(question_div)):
    question = question_div[i].find('a').text
    question_list.append(question)
```

```python
vote_div = soup.find_all('div',class_="s-post-summary--stats js-post-summary-stats")
vote_list = []
for i in range(len(vote_div)):
    try:

        vote = vote_div[i].find('span',class_='s-post-summary--stats-item-number')
        vote_list.append(vote.text)
    
    except:
        vote_list.append(0)

vote_list
```

```python
# Create a dictionary that shows the vote number for each question

result_dict= dict(zip(question_list,vote_list))

result_dict
```

```python
browser.quit()
```

```python
# Visiting first 10 pagies and extracting the questions and votes info

question_list = []
vote_list = []
browser = Browser('chrome')

for i in range(1,5):
    url_1 = f'https://stackoverflow.com/questions/tagged/python?tab=votes&page={i}&pagesize=15'
    browser = Browser('chrome')

    browser.visit(url_1)
    html = browser.html
    soup=BeautifulSoup(html,'html.parser')
    question_div = soup.find_all('div',class_="s-post-summary--content")
    vote_div = soup.find_all('div',class_="s-post-summary--stats js-post-summary-stats")
    for i in range(len(question_div)):
        question = question_div[i].find('a').text
        question_list.append(question)

        try:

            vote = vote_div[i].find('span',class_='s-post-summary--stats-item-number')
            vote_list.append(vote.text)
    
        except:
            vote_list.append(0)
        
    browser.quit()

```

```python
result_dict_pagies= dict(zip(question_list,vote_list))

result_dict_pagies
```
